```python
# Loading the required python modules
import pytesseract # this is tesseract module
import matplotlib.pyplot as plt
import cv2 # this is opencv module
import glob
import os
import pytesseract #import thư viện pytesseract
pytesseract.pytesseract.tesseract_cmd ="C:\\Program Files\\Tesseract-OCR\\tesseract.exe" # Path to tesseract.exe
from deeplearning import OCR, object_detection #import hàm OCR từ file deeplearning.py

# import pandas as pd #thư viện pandas dùng để đọc và xử lý dữ liệu


# specify path to the license plate images folder as shown below
#Lấy tất cả ảnh jph hoặc jpeg, hoặc png trong thư mục
path_for_license_plates = "../2_Train_Object_Detection_model/test_wrong/*"

list_license_plates = []
predicted_license_plates = []
boundingboxs_index = []
# labels_dict = dict(xxmyym = []) #tạo một dictionary chứa các thông tin của các file xml
  
for path_to_license_plate in glob.glob(path_for_license_plates, recursive = True):
      
    license_plate_file = path_to_license_plate.split("/")[-1]
    license_plate, _ = os.path.splitext(license_plate_file)
    #tách dấu cách ở tên file nếu có, chỉ lấy phần trước dấu cách
    license_plate = license_plate.split(" ")[0]
    '''
    Here we append the actual license plate to a list
    '''
    #Thêm vào list_license_plates bắt đầu từ ký tự thứ 9
    list_license_plates.append(license_plate[11:])

    #Lấy ra đường dẫn đến ảnh
    path = path_to_license_plate
    #Lấy ra tên ảnh
    img_name = path.split("/")[-1]
    #Dùng hàm OCR để nhận dạng biển số xe
    filter_predicted_result = OCR(path, img_name)

    #Lấy chỉ số bounding box
    # boudingbox_index = object_detection(path, img_name)
    # boundingboxs_index.append(boudingbox_index)
    # labels_dict["xxmyym"].append(boudingbox_index)
    
    predicted_license_plates.append(filter_predicted_result)
    print('list_license_plates', list_license_plates)
    print('predicted_license_plates', predicted_license_plates)
    # print('boundingboxs_index', boundingboxs_index)
```

```python
print("Actual License Plate", "\t", "Predicted License Plate", "\t", "Accuracy")
print("--------------------", "\t", "-----------------------", "\t", "--------")
  
def calculate_predicted_accuracy(actual_list, predicted_list):
    total_accuracy = 0;
    count_correct = 0;
    for actual_plate, predict_plate in zip(actual_list, predicted_list):
        accuracy = "0 %"
        num_matches = 0
        #Nếu hai ký tự giống hệt nhau thì accuracy = 100%
        if actual_plate == predict_plate:
            accuracy = "100 %"
            count_correct += 1
        #Nếu predic_plate = '' thì accuracy = 0%
        elif predict_plate == '':
            accuracy = "0 %"
        #Nếu không giống hệt nhau thì tính accuracy = max (số ký tự giống nhau/ tổng số ký tự, độ dài chuỗi con dài nhất của predict_plate có trong actual_plate / độ dài actual_plate)
        else:
            for i in range(len(actual_plate)):
                if actual_plate[i] == predict_plate[i]:
                    num_matches += 1
            acc1 = round((num_matches/len(actual_plate))*100, 2)

            #Tính độ dài chuỗi con dài nhất của predict_plate có trong actual_plate
            max_length = 0
            for i in range(len(actual_plate)):
                for j in range(len(actual_plate)):
                    if actual_plate[i:j] in predict_plate:
                        if len(actual_plate[i:j]) > max_length:
                            max_length = len(actual_plate[i:j])

            # acc2= round((max_length/len(actual_plate))*100, 2)
            acc2= round((max_length/max(len(predict_plate), len(actual_plate)))*100, 2)
            #Nếu độ dài actual_plate và predict_plate khác nhau thì accuracy = acc2, ngược lại thì accuracy = max(acc1, acc2)
            # accuracy = str(max(acc1, acc2)) + " %"
            if len(actual_plate) != len(predict_plate):
                accuracy = str(acc2) + " %"
            else:
                accuracy = str(max(acc1, acc2)) + " %"


        print("     ", actual_plate, "\t\t\t", predict_plate, "\t\t  ", accuracy)
        total_accuracy += float(accuracy[:-2])


    #Tính accuracy trung bình
    average_accuracy = round(total_accuracy/len(actual_list), 2)
    print("\nAverage Accuracy: ", average_accuracy, "%")
    #Tính tỉ lệ dự đoán đúng
    correct_rate = round((count_correct/len(actual_list))*100, 2)
    print("Correct Rate: ", correct_rate, "%")



          
calculate_predicted_accuracy(list_license_plates, predicted_license_plates)
```
