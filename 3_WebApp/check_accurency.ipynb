```python
print("Actual License Plate", "\t", "Predicted License Plate", "\t", "Accuracy")
print("--------------------", "\t", "-----------------------", "\t", "--------")
  
def calculate_predicted_accuracy(actual_list, predicted_list):
    total_accuracy = 0;
    count_correct = 0;
    for actual_plate, predict_plate in zip(actual_list, predicted_list):
        accuracy = "0 %"
        num_matches = 0
        #Nu hai k t ging ht nhau th accuracy = 100%
        if actual_plate == predict_plate:
            accuracy = "100 %"
            count_correct += 1
        #Nu predic_plate = '' th accuracy = 0%
        elif predict_plate == '':
            accuracy = "0 %"
        #Nu khng ging ht nhau th tnh accuracy = max (s k t ging nhau/ tng s k t,  di chui con di nht ca predict_plate c trong actual_plate /  di actual_plate)
        else:
            for i in range(len(actual_plate)):
                if actual_plate[i] == predict_plate[i]:
                    num_matches += 1
            acc1 = round((num_matches/len(actual_plate))*100, 2)

            #Tnh  di chui con di nht ca predict_plate c trong actual_plate
            max_length = 0
            for i in range(len(actual_plate)):
                for j in range(len(actual_plate)):
                    if actual_plate[i:j] in predict_plate:
                        if len(actual_plate[i:j]) > max_length:
                            max_length = len(actual_plate[i:j])

            # acc2= round((max_length/len(actual_plate))*100, 2)
            acc2= round((max_length/max(len(predict_plate), len(actual_plate)))*100, 2)
            #Nu  di actual_plate v predict_plate khc nhau th accuracy = acc2, ngc li th accuracy = max(acc1, acc2)
            # accuracy = str(max(acc1, acc2)) + " %"
            if len(actual_plate) != len(predict_plate):
                accuracy = str(acc2) + " %"
            else:
                accuracy = str(max(acc1, acc2)) + " %"


        print("     ", actual_plate, "\t\t\t", predict_plate, "\t\t  ", accuracy)
        total_accuracy += float(accuracy[:-2])


    #Tnh accuracy trung bnh
    average_accuracy = round(total_accuracy/len(actual_list), 2)
    print("\nAverage Accuracy: ", average_accuracy, "%")
    #Tnh t l d on ng
    correct_rate = round((count_correct/len(actual_list))*100, 2)
    print("Correct Rate: ", correct_rate, "%")



          
calculate_predicted_accuracy(list_license_plates, predicted_license_plates)
```
